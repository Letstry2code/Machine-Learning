```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url='https://raw.githubusercontent.com/digipodium/Datasets/main/sample_data.csv'
df=pd.read_csv(url)
df.head()
```

## How to drop columns

```python
df.drop(columns=['country'])   ## Fake Drop, add inplace = True to drop actually
```

Values to add,,, not used much,, add operation

```python
val_to_add=[22,3,2,1,2,4,52,2,2,22,3]
df['new_col']=val_to_add
df
```

## for applying any kind of operation

```python

df['salary_reduced']=df.salary.apply(lambda v:v//100)
```

```python
df
```

### Imputer for missing data 

```python
from sklearn.impute import SimpleImputer
```

## Creating Object

```python
cols_to_impute=['salary','age','salary_reduced']
si=SimpleImputer()
df[cols_to_impute]=si.fit_transform(df[cols_to_impute])
df
```
